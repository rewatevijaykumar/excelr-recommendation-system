```python
import pandas as pd
import numpy as np
```

```python
book = pd.read_csv('dataset/book.csv', encoding='latin1')
book.head()
```

```python
book.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
book
```

```python
book.rename(columns = {'User.ID':'userId', 'Book.Title':'title','Book.Rating':'rating'}, inplace = True)
```

```python
book.head()
```

```python
book.info()
```

```python
#number of unique users in dataset
len(book.userId.unique())
```

```python
#number of unique titles in dataset
len(book.title.unique())
```

```python
#Reshape data (produce a “pivot” table) based on column values.
user_book = book.pivot_table(index='userId', columns='title', values='rating'). reset_index(drop=True)
user_book
```

```python
#Set the index to user ids 
user_book.index = book.userId.unique()
user_book
```

```python
# Impute those NaNs with 0 values
user_book.fillna(0, inplace=True)
user_book
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances( user_book.values,metric='cosine')
user_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
```

```python
#Set the index and column names to user ids 
user_sim_df.index = book.userId.unique()
user_sim_df.columns = book.userId.unique()
```

```python
user_sim_df.iloc[0:5, 0:5]
```

```python
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:5, 0:5]
```

```python
#Most Similar Users
user_sim_df.idxmax(axis=1)[-5:]
```
